2.LOAD DATA

```python
import pandas as pd
df = pd.read_csv('genap.csv', sep=';', engine='python')
df = df.iloc[:, -1]
df = pd.DataFrame(df, columns=['Data'])
df = df.dropna()
df.head()
```

3.TEXT CLEANING

```python
def text_cleaning(text):
    text = str(text)
    text = re.sub(r'http\S+', '', text)
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df['cleaning'] = df['Data'].apply(text_cleaning)
```

```python
nltk.download('punkt_tab')
df['tokenization'] = df['cleaning'].apply(word_tokenize)
```

4. Tokenization

```python
df['tokenization'] = df['cleaning'].apply(word_tokenize)
```

🔸 5. Normalization

```python
normalisasi = {
    "ga": "tidak",
    "gak": "tidak",
    "nggak": "tidak",
    "tdk": "tidak",
    "yg": "yang"
}

def normalization(tokens):
    return [normalisasi.get(word, word) for word in tokens]

df['normalization'] = df['tokenization'].apply(normalization)
```

🔸 6. Stopword Removal

```python
stop_words = set(stopwords.words('indonesian'))

def stopword_removal(tokens):
    return [word for word in tokens if word not in stop_words]

df['stopword'] = df['normalization'].apply(stopword_removal)
```

🔸 7. Case Folding

```python
def case_folding(tokens):
    return [word.lower() for word in tokens]

df['case_folding'] = df['stopword'].apply(case_folding)
```
